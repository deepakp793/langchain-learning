```python
from youtube_transcript_api import YouTubeTranscriptApi

ytt = YouTubeTranscriptApi()

video_id = 'YVBeVM_fA4k'

transscript = ytt.fetch(video_id,("en",))

query = 'Give me complete roadmap disussed in this.'

# complete_transcript : str = ""
# for snippets in transscript:
#     complete_transcript = complete_transcript + " " +snippets.text

complete_transcript = " ".join(snippets.text for snippets in transscript)
```

```python
complete_transcript
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size = 1000,
    chunk_overlap = 20 
)

docs = splitter.create_documents([complete_transcript])
```

```python
len(docs)
docs
```

```python
from langchain_classic.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings

embeddingmodel = OpenAIEmbeddings()

vectorstore = FAISS.from_documents(docs,embeddingmodel)
```

```python
vectorstore.index_to_docstore_id
```

```python
#query = 'What tool is used as frontent?'
```

```python
retriever = vectorstore.as_retriever(search_kwargs = {"k":10})
```

```python
retrieved_doc = retriever.invoke(query)
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate

llm = ChatOpenAI()

template = PromptTemplate(
    template= """
    Act as helpful assistant and provide the information on below query based on provided context below. If information is not sufficient then say I dont know.
    {context},
    question : {query}
    """,
    input_variables=['context','query']
)
```

```python
context = "\n\n".join(doc.page_content for doc in retrieved_doc)
context
```

```python
# Preview the rendered prompt for debugging (don’t overwrite `template`)
prompt_preview = template.invoke({'query': query, 'context': context})
prompt_preview
```

```python
result = llm.invoke(prompt_preview)
result.content
```

#### LCEL

```python
from langchain_core.runnables import RunnableParallel, RunnablePassthrough, RunnableLambda

#query = 'What modules get covered in this video'

def generate_context(retrieved_documents):
    return "/n".join(doc.page_content for doc in retrieved_documents)
    

parallel_chain = RunnableParallel({
    'context' : retriever | RunnableLambda(generate_context),
    'query' : RunnablePassthrough()
})
```

```python
from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()

chain = parallel_chain | template | llm | parser

chain.invoke(query)
```
